```python
import pandas as pd 
import numpy as np 

df_vendas= pd.read_csv("vendas_tech.csv", low_memory="")
display(df_vendas)
df_gerente =pd.read_excel("gerentes_lojas.xlsx") 
display(df_gerente)
```

```python
# Tratamento de dados


# columns
display(df_vendas["Loja"])

# Sempre ter 2 [] quando for passar mais de uma coluna
display(df_vendas[["Loja","Cliente"]])

# Para criar uma nova tabela com apenas colunas
# especificas de uma outra tabela, é possivel usar:

df_novo = (df_vendas[["Loja","Cliente"]])
display(df_novo)

# Excluindo colunas
# Criando uma nova tabela com informações da tabela original,
# excluindo apenas coluna especifica
df_analise = df_vendas.drop(columns=["Data_Base"])

# nulos
# dropna() exclui todas as linhas que tiverem pelo menos um valor vazio
#df_analise = df_analise.dropna()

# Preenchendo valores vazios
df_analise["Loja"] = df_analise["Loja"].fillna("Online")

# tipos de dados
# Alterando tipo de dado na coluna data
df_analise["Data"] = pd.to_datetime(df_analise["Data"],format = "%Y-%m-%d")

# padronização

# str.trip tira todos espaços duplicados, espaços vazios depois do texto
# e espaços vazios antes do texto
df_analise ["Loja"] = df_analise["Loja"].str.strip()
# str.title Faz com que a primeira letra de todas as palavras da coluna fiquem maiusculas
df_analise ["Loja"] = df_analise["Loja"].str.title()

df_gerente ["Loja"] = df_gerente["Loja"].str.strip()
# str.title Faz com que a primeira letra de todas as palavras da coluna fiquem maiusculas
df_gerente ["Loja"] = df_gerente["Loja"].str.title()
# encadeamento
# df_analise["Loja"] = df_analise["Loja"].fillna("Online").str.strip().str.title()


# duplicatas

# drop_duplicates remove as duplicatas
# subset eu digo qual coluna ele deve verificar as duplicatas, podendo ser um conjunto de colunas
# keep "first" = ele mantem o primeiro valor duplicado
# keep "last" = mantem o ultimo valor duplicado exculindo o primeiro
df_analise = df_analise.drop_duplicates(subset=["ID_Pedido"],keep="first")


display(df_analise)
display(df_analise.info())
```

```python
# criar novas colunas

# faturamento
df_analise["Faturamento"] = df_analise["Qtd"] * df_analise["Preco_Unitario"]

# forma de venda
df_analise["Forma_de_Venda"] = np.where(df_analise["Loja"] =="Online","Online","Presencial" )


# regiao
#Primeira forma de fazer
#display(df_analise["Loja"].unique())
#sudeste = ["São Paulo","Rio de Janeiro","Belo Horizonte"]
# isin quer dizer se os valores da coluna loja esta dentro da variavel sudeste.
#df_analise["Regiao"] = np.where(df_analise["Loja"].isin(sudeste), "Sudeste", np.where())

# Segunda forma de fazer regiao

dic_regioes = {
    'São Paulo': "Sudeste",
    'Belo Horizonte':"Sudeste" ,
    'Online': "Online",
    'Rio De Janeiro': "Sudeste",
    'Salvador': "Salvador",
    'Recife': "Nordeste",
    'Curitiba': "Sul",
    'Porto Alegre': "Sul"

}


df_analise["Regiao"] = df_analise ["Loja"].map(dic_regioes)

# Usar função para mapeamento complexo
# def definir_regiao(loja):
#    dic_regioes = {
#   'São Paulo': "Sudeste",
#    'Belo Horizonte':"Sudeste" ,
#    'Online': "Online",
#    'Rio De Janeiro': "Sudeste",
#   'Salvador': "Salvador",
#    'Recife': "Nordeste",
#    'Curitiba': "Sul",
#    'Porto Alegre': "Sul"

# }
#   return dic_regioes(loja)


display(df_analise)
display(df_analise.info())
# isna verifica se ha algum valor vazio em cada coluna
# e o sum vem para somar cada valor vazio que tiver na coluna
display(df_analise.isna().sum())
```

```python
# analisar --> filtrar

# Ordenando a base de dados
# Podendo ordenar por mais de uma coluna
# ascending diz se ordena de maneira ascendente ou descendente com true ou false
df_analise = df_analise.sort_values(by =["Data","Faturamento"],  ascending = False)
df_analise = df_analise.reset_index(drop=True)
# .loc: id do pedido -- localizando informações de uma linha especifica
# atraves do indice que é o nome da linha e da coluna
id_pedido = 4
loja = df_analise.loc[df_analise["ID_Pedido"]==4,"Loja"].values[0]
produto = df_analise.loc[df_analise["ID_Pedido"]==4,"Produto"].values[0]
cliente = df_analise.loc[df_analise["ID_Pedido"]==4,"Cliente"].values[0]
print(loja)
print(produto)
print(cliente)


# .iloc -> localiza por posicao
id_pedido = 4
loja = df_analise.iloc[3,2]
produto = df_analise.iloc[3,3]
cliente = df_analise.iloc[3,6]
print(loja,produto,cliente)

#condicional

df_id_pedido_4 = df_analise[df_analise["ID_Pedido"]==4]

# exportar pedaços da base

df_vendas_sp = df_analise[df_analise["Loja"]=="São Paulo"]
df_vendas_sp.to_csv("Vendas_SP.csv",index = False)

# exportar as vendas de 2024

df_vendas_2024 = df_analise[df_analise["Data"].dt.year==2024]

# duplas condições
# Se caso eu quisesse aplicar o or usaria | 
# Ja o and é &
df_vendas_HDMI_sul = df_analise[(df_analise["Produto"] == "Cabo HDMI") & (df_analise["Regiao"] == "Sul")]

display(df_vendas_HDMI_sul)

display(df_vendas_2024)

display(df_analise)
display(df_id_pedido_4)
display(df_vendas_sp)
```

```python
# analise por agrupamento
#display(df_analise)

# analise de ranking por loja
# Fazendo com q a coluna loja nao se torne um index com o as_index
analise_lojas = df_analise[["Loja","Faturamento"]].groupby("Loja",as_index=False).sum()
analise_lojas = analise_lojas.sort_values(by="Faturamento",ascending = False)
# obs: ao formatar os valores da coluna se tornam string ou seja
# nao tem como mais usar os valores para operações matematicas
analise_lojas["Faturamento"] = analise_lojas["Faturamento"].map("R${:,.2f}".format)
display(analise_lojas)

# Ranking de produtos que mais venderam no online
df_vendas_online = df_analise[df_analise["Loja"] == "Online"]
analise_produtos_online = df_vendas_online[["Produto", "Qtd"]].groupby("Produto",as_index=False).sum()
analise_produtos_online = analise_produtos_online.sort_values(by="Qtd",ascending=False)
# Mudar nome de uma coluna
analise_produtos_online = analise_produtos_online.rename(columns={"Qtd":"Vendas Totais"})
display(analise_produtos_online)

# analise de ranking por loja e por produto
# quais produtos venderam mais em cada uma das lojas
analise_produtos_lojas = df_analise[["Loja","Produto","Qtd"]].groupby(["Loja","Produto"]).sum()
#Tirando o limite de linha do display em um contexto especifico
# ou seja so para analisar a tabela analise_produtos_lojas

# quais lojas mais venderam os produtos
analise_lojas_produtos = df_analise[["Loja","Produto","Qtd"]].groupby(["Produto","Loja"]).sum()

display(analise_lojas_produtos)

with pd.option_context("display.max_rows", None):
    display(analise_produtos_lojas)








```

```python
display(df_analise.head())

# quais gerentes bateram a meta em janeiro de 2023

df_meta = df_analise[(df_analise["Data"].dt.year == 2023) & (df_analise["Data"].dt.month == 1)]

df_meta = df_meta[["Loja","Faturamento"]].groupby("Loja",as_index=False).sum()

df_meta = df_meta.merge(df_gerente, on="Loja", how="left")
df_meta["Bateu Meta"] = np.where(df_meta["Faturamento"] >= df_meta["Meta_Mensal"], "Sim","Não")


display(df_meta)
```
